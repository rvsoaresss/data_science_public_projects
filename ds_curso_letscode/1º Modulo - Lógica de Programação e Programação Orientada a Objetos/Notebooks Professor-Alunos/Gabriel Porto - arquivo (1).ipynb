```python
lista_de_listas = [] 
```

```python
for i in range(len(lista1)):  #transformando em lista de listas
    lista_de_listas.append(lista1[i].split(','))
```

```python
def imprimir_linhas(n,m): #mostrar linhas de n a m
    for i in range(n,m+1):
        print(lista_de_listas[i])
```

```python
imprimir_linhas(4,7)
```

```python
def imprimir_nome(nome): #imprimir a linha com o nome procurado
    #print(lista_de_listas)
    for i in range(len(lista_de_listas)):
        for j in range(len(lista_de_listas[i])):
            if lista_de_listas[i][j] == nome:
                print(lista_de_listas[i])
                
            
```

```python
imprimir_nome('Betina')
```
