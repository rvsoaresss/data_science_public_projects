```python
class Restaurante: 
    def __init__ (self, esquerda, direita, atributo):
        self.esquerda = esquerda
        self.direita = direita
        self.atributo = atributo
        
    def sugerir_prato(self):
        
        if self.direita == None and self.esquerda == None:
            print('O prato sugerido é',self.atributo)
        else:
            print(self.atributo)
            resposta = input('\tDigite "s" para sim e "n" para não: ')
            if resposta == 's':
                self.esquerda.sugerir_prato()
            elif resposta == 'n':
                self.direita.sugerir_prato()
            else:
                print('\tDigite a opção correta!')
                self.sugerir_prato()
        
    
```

```python
mousse = Restaurante(esquerda = None,direita = None,atributo ='Mousse Diet')
petit = Restaurante(esquerda = None,direita = None,atributo ='Petit Gateau')
file = Restaurante(esquerda = None,direita = None,atributo ='Filet Mignon')
abacaxi = Restaurante(esquerda = None,direita = None,atributo ='Abacaxi')
tofu = Restaurante(esquerda = None,direita = None,atributo ='Tofu')
acucar = Restaurante(esquerda = petit,direita = mousse,atributo ='Pode comer açúcar?')
prato_principal_carnivoro = Restaurante(esquerda = file,direita = acucar,
                                        atributo ='Prato principal?')
prato_principal_vegano = Restaurante(esquerda = tofu,direita = abacaxi,
                                     atributo ='Prato principal?')
vegano = Restaurante(esquerda = prato_principal_vegano,direita = prato_principal_carnivoro,
                     atributo ='É vegano?')

```

```python
vegano.sugerir_prato()
```
