pediu que fosse feito um software para que os diretores de venda pudessem simular como
seria a venda de outros produtos. Para viabilizar o projeto, a Routhon disponibilizou uma
base de dados com alguns produtos, suas características e se a venda desse produto foi alta
ou não. Vocês, como consultores da PyCoders Ltda., foram designados a desenvolver esse
case. Com a base de dados em mãos, é necessário identificar os padrões que fazem um
produto ser muito vendido ou não e, em seguida, desenvolver o software para que os

```python
class Loja:
    def __init__(self,atributo,esquerda = None ,direita = None):
        self.esquerda = esquerda
        self.direita = direita
        self.atributo = atributo
        
    def analisar(self):
        if self.direita == None and self.esquerda == None:
            print('O produto tem boas vendas?:',self.atributo)
        else:
            print(self.atributo)
            resposta = input('\tDigite "s" para sim e "n" para não: ')
            if resposta == 's':
                self.esquerda.analisar()
            elif resposta == 'n':
                self.direita.analisar()
            else:
                print('\tDigite a opção correta')
                self.analisar()
```

```python
nao_vende_crianca = Loja('Não',esquerda = None, direita = None)
vende_crianca = Loja('Sim',esquerda = None, direita = None)
vende_adulto_inv  = Loja('Sim',esquerda = None, direita = None)
vende_adulto_verao = Loja('Sim',esquerda = None, direita = None)
nao_vende_adulto_verao = Loja('Não',esquerda = None, direita = None)

e_baixo_crianca = Loja('O preço é baixo?',esquerda = vende_crianca, direita = nao_vende_crianca)
e_alto_adulto = Loja('O preço é alto?',esquerda = nao_vende_adulto_verao,direita = vende_adulto_verao)
verao_adulto = Loja('É para o verão?',esquerda = e_alto_adulto, direita = vende_adulto_inv)
novo_produto = Loja('É para adulto?',esquerda = verao_adulto, direita = e_baixo_crianca)

```

```python
novo_produto.analisar()
```
