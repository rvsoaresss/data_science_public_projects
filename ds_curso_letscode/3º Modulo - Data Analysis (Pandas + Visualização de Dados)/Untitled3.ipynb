```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("WineDataset.csv", sep="|", decimal=".")
```

```python
df
```

```python
target=df["target"]
```

```python
target=target.replace({0: "class_0"})
target=target.replace({1: "class_1"})
target=target.replace({2: "class_2"})
df["target"]=target
```

```python
filtro = df['target']=='class_0'
plt.scatter(df[filtro]['proline'],df[filtro]['od280/od315_of_diluted_wines'],label='class_0',marker='x')

filtro = df['target']=='class_1'
plt.scatter(df[filtro]['proline'],df[filtro]['od280/od315_of_diluted_wines'],label='class_1',marker='o')

filtro = df['target']=='class_2'
plt.scatter(df[filtro]['proline'],df[filtro]['od280/od315_of_diluted_wines'],label='class_2',marker='^')

plt.title('Desafio')
plt.legend(loc='upper right')
plt.xlabel('od280/od315_of_diluted_wines')
plt.ylabel('proline')
plt.show()
```

```python
def modelo(proline, diluted_wines):
    if proline >= 2.5 and diluted_wines >=750:
        return ("class_0")  

    elif diluted_wines <750:
        return ("class_1")
    
    elif proline < 2 :
        return ("class_2")
    else:
        return ("class_2")
        

prev = []

for i in range(df.shape[0]):
    target = modelo(df['proline'].iloc[i],
                               df['od280/od315_of_diluted_wines'].iloc[i])
    prev.append(target)
    
df['prev'] = prev

sum(df['target']==df['prev'])/df.shape[0]
```
