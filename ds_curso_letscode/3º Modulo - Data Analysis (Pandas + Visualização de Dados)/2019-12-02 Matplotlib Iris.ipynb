```python
mat = pd.Series([2,7,5,10,6])
mat.index = ['Romario','Pedro','José','Diego','Abraão']


pt = pd.Series([5,5,3,9,2])
pt.index = ['Romario','Pedro','José','Diego','Abraão']

plt.bar(mat.index, mat, label="Matematica")
plt.bar(pt.index, pt, label="Portugues")

plt.legend()

plt.show()
```

```python
setosa_sepal = df[df['species'] == 'setosa']['sepal_width']
setosa_petal = df[df['species']== 'setosa']['petal_width']
virginica_sepal = df[df['species'] == 'virginica']['sepal_width']
virginica_petal = df[df['species']== 'virginica']['petal_width']
versicolor_sepal = df[df['species'] == 'versicolor']['sepal_width']
versicolor_petal = df[df['species']== 'versicolor']['petal_width']
plt.scatter(setosa_petal,setosa_sepal,color='blue',marker='x')
plt.scatter(virginica_petal,virginica_sepal,color='orange',marker='o')
plt.scatter(versicolor_petal,versicolor_sepal,color='green',marker='^')
plt.legend(['setosa','virginica','versicolor'])
plt.title("Desafio")
plt.xlabel("petal_width")
plt.ylabel("sepal_width")
plt.savefig("Desafio.png")
plt.show()
```
