Importar os dois arquivo banks e juntar em apenas um DataFrame

```python
df = pd.read_excel('banks.xlsx',sheet_name='Sheet1')
df
```

```python
lista = df['Acquiring Institution'].str.split(', ')
lista2 = []
contador = []
for i in range(len(lista)):
    if 'National Association' in lista.iloc[i]:
        lista.iloc[i].remove('National Association')
    if 'N.A.' in lista.iloc[i]:
        lista.iloc[i].remove('N.A.')
    if 'Inc.' in lista.iloc[i]:
        lista.iloc[i].remove('Inc.')
    contador.append(len(lista.iloc[i]))
    string = lista.iloc[i][0]
    for j in range(1,len(lista.iloc[i])):
        string = string +', '+lista.iloc[i][j]
    lista2.append(string)
df['Acquiring Institution'] = lista2
df['Acquiring Institution Count'] = contador
```

```python
df.loc[df['Acquiring Institution']=='NO ACQUIRER', 'Acquiring Institution'] = np.nan
```

```python
lista = df[df['Acquiring Institution Count']<2]['Acquiring Institution'].value_counts()
for linha in df[df['Acquiring Institution Count']>1]['Acquiring Institution']:
    for elem in linha.split(', '):
        if elem in lista.index:
            lista.loc[lista.index==elem] = lista.loc[lista.index==elem].values[0]+1
        else:
            helper = pd.Series(data=[1],index = [elem])
            lista = lista.append(helper)
lista.sort_values(ascending=False)[:10]
```
